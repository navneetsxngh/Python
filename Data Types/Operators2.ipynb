#### Identity Operator
It checks whether the ID of two different data are same or not.
Data and data type should be immutable.

```python
a = 10
b = 10
print(a is b)
print(id(a))
print(id(b))
```

```python
c = [1,2]
d = [1,2]
print(id(c))
print(id(d))
print(c is d)
```

```python
s1 = {1,2}
s2 = {1,2}
print(s1 is not s2)
```

```python
t1 = (1,2)
t2 = (1,2)
print(type(t1))
print(type(t2))
print(id(t1))
print(id(t2))
print(t1 is t2)
```
