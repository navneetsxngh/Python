## TUPLE

```python
# (1). Identification - ()
# (2). Immutable datatype ---> We can't change or update values at a run time.
# (3). Indexing starts from 0 and length starts from 1.
# (4). This is like a container in which we can store different-different data type's elements.
```

```python
a = ('sam', 'raj', 56, 23.89)
print(a)
print(type(a))
```

```python
## Indexing in Tuple
a = ('sam', 'raj', 56, 23.89)
print(a[0])
print(a[1])
print(a[2])
```

```python
## Slicing in Tuple
a = ('sam', 'raj', 56, 23.89)
print(a[0:3])
print(a[1:4])
print(a[2:])
print(a[:2])
print(a[:])
```

```python
## Immutable Datatype
a = ('sam', 'raj', 56, 23.89)
# a[1] = 'rahul'  ----> Error at a runtime.
```

```python
## Properties of Tuple
# (1). Concatenation
a = (1,2,3,4)
b = (4,5,6,7)
c = a + b
print("c: ", c)

# (2). Replication 
d = (3,4,5)
e = d*3
print("e: ", e)
```

```python
# Functions in Tuple
# (1). len() ---> It will return length of a tuple.
a = ('sam', 'raj', 56, 23.89)
print(len(a))
```

```python
# (2). index() ---> It will return index number of a tuple.
a = ('sam', 'raj', 56, 23.89)
a.index(56)
```

```python
# (3). count() ---> It will return the frequency of a element in a tuple.
a = ('sam', 'raj', 56, 23.89,56,56,56,56,56)
a.count(56)
```

```python
# (4). min() ---> It will return the minimum number in a tuple.
a = (1,2,3,4)
min(a)
```

```python
# (5). max() ---> It will return the maximum number in a tuple.
a = (4,5,6,7)
max(a)
```

```python
#Q1. Reverse a Tuple without using slicing.?
tup = []
size = int(input("Enter the size of the tuple: "))

for i in range(size):
    val = input("Enter Items")
    if val.isdigit():
        tup.append(int(val))
    else:
        tup.append(val)
print("Tuple: ", tuple(tup))
rev = []
for i in range(1,len(tup)+1):
    rev.append(tup[-i])

print("Reversed: ", tuple(rev))
```

```python
#Q2. Find the second largest number in a tuple.?
tup = []
size = int(input("Enter the size of the tuple: "))

for i in range(size):
    val = input("Enter Items")
    if val.isdigit():
        tup.append(int(val))
    else:
        tup.append(val)
print("Tuple: ", tuple(tup))
max = 0
max2 = 0
for i in tup:
    if i > max:
        max = i
for i in tup:
    if i > max2 and i < max:
        max2 = i
print("The Second Largest Number in Tuple is: ", max2)
```

```python
#Q8. Create a tuple of Unique characters from a string.
a = 'programming'
c = ''
for i in a:
    b = i + " "
    c += b
d = c.split()

e = []
for i in d:
    if i not in e:
        e.append(i)
print(e)
```

```python
#Q10 . Generate all possible pairs from two tuples. 
a = (1,2)
b = ('a', 'b')

c = [*((i,j) for i in a for j in b),]
print(tuple(c))
```

```python
#Q11. Group elements of a tuple into pairs.
a = (1, 2, 3, 4, 5, 6)
b = []

for i in range(0, len(a), 2): 
    b.append((a[i], a[i+1]))
print(b)
```

```python
# a = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
# output =[(2, 1), (1, 2), (2, 3), (4, 4), (2, 5)]
# last element of each tuple should be in ascending order

a = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
b = a[:]

for i in range(len(b)):
    for j in range(len(b) - 1):
        if b[j][1] > b[j + 1][1]:
            b[j], b[j + 1] = b[j + 1], b[j]

print(b)
```

```python
#Q12. Create a tuple of square from 1 to 4 using comprehension?

a = tuple(i**2 for i in range(1,5))
print(a)
a = tuple(i**2 for i in range(1,10) if i % 2 == 0)
print(a)

b = tuple("Even" if x % 2 == 0 else x for x in range(1,5))
print(b)
```

```python
#Q13. From the string 'Python' create a tuple that convert vowels to uppercase and consonants in lowercase.
a = 'Python'
vowels = 'aeiou'

b = tuple(i.upper() if i.lower() in vowels else i.lower() for i in a)
print(b)
```
