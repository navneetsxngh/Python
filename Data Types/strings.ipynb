```python
## Strings in Python

#(1). Identification = '', " ", ''' '''
#(2). Immutable Data Type ----> We cannot change or update values at a run time.
#(3). Indexing start from 0 and length start from 1.
#(4). Forward Indexing start from 0 and backward indexing start from -1.

a = "Python"
print(a[0])    ### Indexing
print(a[1])
print(a[4])
```

```python
### SLicing In Strings
#[start_index:stop_index:step_value]

b = "I Love Python"
print(b[0:6])
print(b[2:])
print(b[:])
```

```python
## Properties of String
# (1). Concatenation

c = '12' + '34'
print(c)

# (2). Replication
d = "sam" * 3
print(d)
```

```python
## Function in a String
# (1). len() ---> It will return length of a string
a = "sam"
len(a)
```

```python
# (2). Index()----> It will return indexing of a string.
b = "Python"
b.index('o')
```

```python
# (3). count()---> It will return frequency of element in a string.
a = "aloooo"
a.count('o')
```

```python
# (4). isalpha() ---> It will return true if there is only alphabet in a string.
#Otherwise it will return false.

a = "abc"
a.isalpha()
```

```python
# (5). isdigit() ---> It will return True if there is only digits in a string.
# Otherwise it will return False.

c = '123456'
c.isdigit()
```

```python
# (6). isalnum() ----> It will return True if there is either alphabet or digits or both alphabets and digits.
d = "sam123"
d.isalnum()
```

```python
# (7). lower() ---> It will convert all charcaters into lowercase.
e = "PYTHon"
e.lower()
```

```python
# (8). upper() ---> It will convert all characters into uppercase.
f = 'pythoN'
f.upper()
```

```python
# (9). ---> replace(): Replace a substring with another string.
text = "I love Python"
print(text.replace("Python", "Coding"))
```

```python
# (10). split(): Splits a string into a list.
text = "I Love Python"
print(text.split(","))
```

```python
# (11). join(): Joins elements of a list into a single string.
fruits = ['apple', 'banana', 'cherry']
print(", " .join(fruits))
```

```python
### String Formatting
name = 'Rohit'
age = 30
print(f"My name is {name} and i am {age} years old.")
```

```python
# (2). Using format():
name = "Raj"
age = 25
print("My name is {} and i am {} years old." .format(name,age))
```
