```python
# Bakery shop ---> 
# Display('Burgur':50 , 'Pizza':100 , 'Sandwich':200) 
# user ---> Item ---> Quantity ---> User 500 > price  ----> 10% discount , 1000 > price 20% discount .
```

```python
print("Here's our Menu")
items = ['Burger', 'Pizza', 'Sandwich']
print(items)
amount = 0

for item in range(1,13):
    item = input("Enter the item you like to Purchase: ").lower()
    if item == 'burger':
        print("Burger cost = 50")
        quantity = int(input("Enter Quantity: "))
        cost = 50 * quantity
    elif item == 'pizza':
        print("Pizza Cost = 100")
        quantity = int(input("Enter Quantity: "))
        cost = 100 * quantity
    elif item == 'sandwich':
        print("Sandwich cost = 200")
        quantity = int(input("Enter Quantity: "))
        cost = 200 * quantity

    amount += cost
    
    wish = input("Would you like to purchase anything else (y/n): ").lower()
    if wish == 'y':
        continue
    if wish == 'n':
        break

print(f"Amount to be Paid: {amount}")

if amount > 500 and amount <= 1000:
    print("10% Discount")
    discount = 0.1 * amount
    effamount = amount - discount
    print(f"Effective amount to be Paid: {effamount}")
elif amount > 1000:
    print("20% Discount")
    discount = 0.2 * amount
    effamount = amount - discount
    print(f"Effective amount to be paid: {effamount}")
else:
    print("No Discount")

print("Thank You for your Purchase")
```
