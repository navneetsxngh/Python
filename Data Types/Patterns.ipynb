### Square hollow Pattern

```python

n = 5
for i in range(1,n+1):
    for j in range(1,n+1):
        if j == 1 or i == 1 or i == n or j == n:
            print("*", end= " ")
        else:
            print(" ", end= " ")
    else:
        print()
```

### Number Triangular

```python
n = 5
for i in range(1,n+1):
    print(" " * (n - i), end= " ")
    for j in range(i):
        print(i, end=" ")
    else:
        print()
```

### Number Increasing Pyramid

```python
n = 5
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

### Number Increasing Reverse Pattern

```python
n = 5
for i in range(n,0,-1):
    for j in range(1,i+1):
        print(j, end=" ")
    else:
        print()
```

### Number Changing Pyramid

```python
n = 5
temp = 1
for i in range(1,n+1):
    for j in range(1,i+1):
        print(temp, end=" ")
        temp += 1
    else:
        print()
```

### Diamond Pattern

```python
n = 5
for i in range(n-1):
    for j in range(i+1,n):
        print(" ", end=" ")
    for j in range(i+1):
        print("*", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
for i in range(n):
    for j in range(i):
        print(" ", end=" ")
    for j in range(i+1,n):
        print("*", end=" ")
    for j in range(i,n):
        print("*", end=" ")
    print()
```

### Left Half Pyramid

```python
n = 5
for i in range(n):
    for j in range(i,n):
        print(" ", end=" ")
    for j in range(i+1):
        print("*", end=" ")
    print()
```

### Reverse Left Half Pyramid

```python
n = 5
for i in range(n):
    for j in range(i):
        print(" ", end=" ")
    for j in range(i,n):
        print("*", end=" ")
    print()
```

### K-Pattern

```python
n = 5
for i in range(n):
    for j in range(i,n):
        print("*", end=" ")
    print()
for i in range(n):
    for j in range(i+1):
        print("*", end=" ")
    print()
```

### Hill Pattern

```python
n = 5
for i in range(n):
    for j in range(i+1,n):
        print(" ", end=" ")
    for j in range(i+1):
        print("*", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
```

### Reverse Hill Pattern

```python
n = 5
for i in range(n):
    for j in range(i):
        print(" ", end=" ")
    for j in range(i+1,n):
        print("*", end=" ")
    for j in range(i,n):
        print("*", end=" ")
    print()
```
