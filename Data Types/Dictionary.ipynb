## Dictionary

```python
#Introduction to Dictionary
#(1). Identification ---> {key:value}
#(2). Element will seperate by comma.
#(3). Indexing start from 0 and length start from 1.
```

```python
a = {'Model': 'Hyarader', 'Brand': 'Toyota', 'Price': 20000000}
print(type(a))
```

```python
#Loop in Dictionary
##Extract all keys 
for i in a:
    print(i)
```

```python
for i in a.keys():
    print(i)
```

```python
##Extract all values
for i in a:
    print(a[i])
```

```python
for i in a.values():
    print(i)
```

```python
### Extract all keys and values
for i,j in a.items():
    print(i,j)
```

```python
##Update the Values
a['Model'] = 'Innova'
a['Colour'] = 'Black'
print(a)
```

```python
## Functions in Dictionary
# (1). length----> It will return length of a dictionary
len(a)
```

```python
# (2). .remove() ---> It will remove key value pair.
a.pop('Colour')
print(a)
```

```python
# (3). popitem() ----> It will remove element from the last index.
a.popitem()
print(a)
a.popitem()
print(a)
```

```python
# (4). update() ----> It will update value of a specific key
a = {'Model': 'Hyarader', 'Brand': 'Toyota', 'Price': 20000000}
a.update({"Model":"Innova"})
print(a)
```

```python
#Q8 Count the frequency of each character in a string using a dictionary.
a = "banana"
#output ---> {"b": 1, "a": 3, "n": 2}

b = {i : a.count(i) for i in a}
print(b)
```

```python
#Q9 Count the frequency of words in a list using dictionary.
words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
#output = {'Apple': 3, "banana": 2, "orange": 1}

c = {i:words.count(i) for i in words}
print(c)
```

## Dictionary Comprehension

```python
# Dictionary Comprehension in python offer a consice way to create dictionaries from iterables.
# Similar to List Comprehension, they allow you to construct dictionaries in a single line, with an optional condition for filtering
# Syntax:
# {Key_Expression : Value_expression for item in iterable if condition} 
```

### Q1 Create a Dictionary of Squares.?

```python
sq = {x:x**2 for x in range(1,11)}
print(sq)
```

### Q2 Filter Even Numbers

```python
even = {x:x**2 for x in range(1,11) if x % 2 == 0}
print(even)
```

### Q3 Reverse a Dictionary

```python
original = {'a': 1, 'b': 2, 'c': 3}
reverse = {k:v for v,k in original.items()}
print(reverse)
```

### Q4 Convert List to Dictionary with Lengths

```python
words = ["apple", "banana", "cherry"]
length_dict = {x : len(x) for x in words}
print(length_dict)
```

### Q5 Create a Dictionary where keys are words and values are uppercase.

```python
words =['python', 'code', 'dict']

t = {x: x.upper() for x in words}
print(t)
```

### Q6 Create a Dictionary from a setence with each word and its frequency


```python
sentence = 'python makes coding easy and python is fun'
b = sentence.split()

e = {x : b.count(x) for x in b}
print(e)
```

### Q7 Create a nested Dictionary of numbers form 1-3 with their square and cube.

```python
numbers = {x: {"Square": x**2, "Cube": x**3} for x in range(1,4)}
print(numbers)
```

### Q8 Create a Dictionary from a string where keys are vowels and values are their counts.

```python
text = "dictionary comprehension practise"
vowels = 'aeiou'

p = {x : text.count(x) for x in text if x in vowels}
print(p)
```

### Q9 Create a Dictionary with numbers 1-10 as keys and "Divisible by 3" / "Not divisible" as values.

```python
t = {x : "Divisible By 3" if x % 3 == 0 else "Not Divisible" for x in range(1,11)}
print(t)
```

### Q10 Create a dictionary from 2 lists

```python
keys = ['name', 'age', 'city']
values = ['sam', 30, 'Delhi']

s = {k:v for k,v in zip(keys,values)}
print(s)
```

### Q11 Convert a list of tuples into a dictionary
