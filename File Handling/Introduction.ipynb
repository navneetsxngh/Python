```python
# Python Program(open, read, write) ---> Buffer Ram ---> Operating System(System call, File Descrictor) ---> File System ----> Disk Controller.
```

```python
# Opening a File.
file = open("file.txt", "r")
file.read()
file.close()
```

```python
f = open("file.txt", "r")
print(f.read())
f.close()
```

```python
## Writable File
file = open("file.txt", "w")
file.write("This is the line from the writing file (write mode)")
file.close()

## This overwrite the file.
```

```python
# If any text is already present into the file then how can we add the next line.
# For this we use append mode ('a').

file = open("file.txt", "a")
file.write("\nThis is a Text Append Mode.")
file.write("\nThis is a New line.")
file.close()
```

```python
sentences = ["This is the First line from the list.", "\nThis is the Second line from the list.", "\nThis is the Third line from the list."]

file = open("sample.txt", "w")
for i in sentences:
    file.write(i)
file.close()
```

```python
# Read, Readline, Readlines
file = open("sample.txt", "r")
print(file.readline())
print(file.readline())
print(file.readline())
print(type(file.read()))

file.close()
```

```python
file = open("sample.txt", "r")
li = file.readlines()
print(li)
print(li[2])
file.close()
```

```python
## It closes the file automatically after I/O operation.
with open("sample.txt", "r") as file:
    print(file.read())
```

```python
## Seek and tell function into the file handling
with open("sample.txt", "r") as f:
    print(f.readline())
    print(f.tell())    #---> It tells the iterator.

    # Seeking the pointor to 10
    print(f.seek(10))   #---> It starts the iterator from the given position.
    
    print(f.readline())
```

```python
## Writelines
# Take big files

big_data = ["Hello World!!! \n" for i in range(1,2000)]

with open("new.txt", 'w') as f:
    f.writelines(big_data)
```

```python
## Reading big files into the small chunks.
with open("new.txt", "r") as f:
    chunk_size = 100
    while True:
        chunk = f.read(chunk_size)

        if not chunk:
            break
        print("100 Words Completed")
        print(chunk)
```

```python
## r+ mode in file Handling
# This is used for the reading and writing the file but it works on the existing file only.

with open("sample.txt", "r+") as f1:
    f1.write("anything")    ## It overwrites the texts.
    f1.seek(0)
    print(f1.read())
```

```python
## w+ mode in file handling
# This mode is used for reading and writing the file and also creates a new file if it doesn't exist

with open("file1.txt", "w+") as f2:
    f2.write("Hello Everyone ")
```

```python
## Reading a CSV(comma seperated values) file using Python

import csv

with open("student_marks.csv", "r") as f1:
    content = csv.reader(f1)
    for line in content:
        print(line)
```
