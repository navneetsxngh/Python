## Q2. Write a program to create a file named hello.txt and write "Hello, Python!" into it.

```python
with open("hello.txt", "w") as f1:
    f1.write("Hello, Python! ")
```

## Q3 Write a program to read the content of hello.txt and print it.

```python
with open("hello.txt", "r") as f2:
    print(f2.read())
```

## Q4 Write a Program to append a new line "Welcome to Python" to hello.txt

```python
with open("hello.txt", "a+") as f2:
    f2.write("\nWelcome to Python")
    f2.seek(0)
    print(f2.read())
```

## Q7 Write a program to copy the contents of "hello.txt" to another file "copy.txt"

```python
with open("hello.txt", "r") as source, open("copy.txt", "w+") as destination:
    copy = source.read()
    paste = destination.write(copy)
    print(destination.read())
```

## Q8 Write a program to write a list of numbers [1,2,3,4,5] into a file, one number per line.


```python
with open("hello.txt", "w+") as f3:
    for i in range(1, 6):
        f3.write(f"{i}\n")
    f3.seek(0)  
    print(f3.read())
```

## Q9 Write a program to read all numbers from a file and calculate their sum.

```python
with open("hello.txt", "r") as f4:
    sum = 0
    for num in f4:
        sum += int(num)
    print("Sum of Numbers in Given file: ", sum)
```
