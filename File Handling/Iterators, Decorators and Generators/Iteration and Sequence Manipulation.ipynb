## 1.) Sorted(iterable, /, *, key = None, Reverse = False)

```python
list1 = [1,12,7,-3,8]
print(sorted(list1))
print(sorted(list1, reverse = True))
print(sorted(list1, key= abs))
```

## 2.) Reversed(seq, /)

```python
list2 = [1,12,7,-3,8]
# reversed(list2)   ## It makes a Object

rev = reversed(list2)
print(list(rev))
```

## 3.) Slice(start, stop, step)

```python
list3 = [10,20,30,40,50,60,70]

s = slice(5)
list3[s]
```

## 4.) iter(callable, sentinel)

```python
list4 = [10,20]
it = iter(list4)
print(next(it))
print(next(it))
```
