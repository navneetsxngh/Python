# Decorators

A decorator is essentially a function that takes another function as argument and returns a new function with enhanced functionality.

```python
def my_decorator(fun):
    def wrapper():
        print("This task will run before the main function runs")
        fun()
        print("This is the Task which will run after the main function runs")
    return wrapper

@my_decorator
def display():
    print("This is another task")
    print("This is the display function")

display()
```

```python
def decorator(a, b):
    def inner_decorator(func):
        def wrapper():
            print("This task will run before the main function runs")
            result = func(a, b)
            print("This is the task which will run after the main function runs")
            return result
        return wrapper
    return inner_decorator

@decorator(2, 3)
def multiply(a, b):
    print(f"The result of multiplying {a} and {b} is {a * b}")

multiply()
```

```python
def decorator(f):
    def inner():
        print("First Line")
        f()
        print("Second Line")
    return inner

@decorator
def display():
    print("Welcome")
display()
```
