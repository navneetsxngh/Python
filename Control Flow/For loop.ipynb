#### Sum of N natural Numbers

```python
num = int(input("Enter limit: "))
sum = 0
i = 0

for i in range(0, num+1):
    sum += i

print("The sum of N natural Numbers: ", sum)
```

#### Factorial of given Number

```python
num1 = int(input("Enter Number: "))
factorial = 1

for j in range(num1,0,-1):
    factorial *= j

print("The Factorial of the given Number is: ", factorial)
```

#### Fibonacci Series

```python
num2 = int(input("Enter Limit: "))
first = 0
second = 1

for k in range(0,num2+1):
    next = first + second
    print(first)
    first = second
    second = next
```

#### Factors of a Number

```python
num3 = int(input("Enter Number: "))

for i in range(1,num3+1):
    if num3 % i == 0:
        print(i)
```

#### Prime Number Check

```python
num4 = int(input("Enter Number: "))
factor = 0

for i in range(2,num4):
    if num4 % i == 0:
        factor += 1

if factor == 0:
    print("Prime Number")
else:
    print("Not a Prime Number")    
```

## Nested Loop

```python
for i in range(1,4):
    for j in range(1,4):
        print(i,',',j, end='  ')
    print(' ')
```

#### Prime Number Between 1 to N

```python
num5 = int(input("Enter Limit: "))

for i in range(2,num5+1):
    count = 0
    for j in range (2,i):
        if i % j == 0:
            count += 1
    if count == 0:
        print(i, end= ' ')
```

#### Square Pattern

```python
n = int(input("Enter Number: "))

for i in range(1,n+1):
    for j in range(1,n+1):
        print('*',end=' ')
    print(' ')
```

### LEft Handed Triangle

```python
n = int(input("Enter limit: "))

for i in range(1,n+1):
    for j in range(1,n+1):
        if i >= j:
            print('*', end=' ')
    print('')
```
