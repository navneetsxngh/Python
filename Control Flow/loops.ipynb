### Decreasing Pattern

```python
n = 5
for i in range(1,n+1):
    temp = 1
    for j in range(i,n+1):
        print(temp, end= " ")
        temp += 1
    print()
```

```python
n = 5
for i in range(1, n+1):
    print(" "*(n-i), end = " ")
    for j in range(1, i+1):
        print(j, end = " ")
    print()
```

#### Hollow Square Pattern

```python
n = 5
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print(j+1, end = " ")
        else:
            print(" ", end = " ")
    print()
```

```python
n = 5
for i in range(1,n+1):
    for j in range(1, i+1):
        if j == 1 or j == i or i == n:
            print('*', end= "")
        else:
            print(" ", end="")
    print()
```

### Left Increasing Triangle

```python
n = 5
for i in range(1,n+1):
    for j in range(i,n):
        print(" ", end=" ")
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

### Left Decreasing Triangle

```python
n = 5
for i in range(1,n+1):
    for j in range(i):
        print(" ", end=" ")
    for j in range(i,n+1):
        print(j, end=" ")
    print()
```

## WHILE Loop

```python
# (2). while --->
# (a). Initialization
# (b). Condition
# (c). Increment or Decrement
```

```python
#Q.1 Write a Program to print 1 to 4?

i = 1
n = 4
while i <= n:   ##Condition
    print(i, end=" " )
    i += 1      ## Increment
```

```python
#Q2. Write a program who print 4 to 1?
i = 4
n = 1
while i >= n:
    print(i)
    i -= 1
```

```python
#Q3. Write a Program to sum the given numbers in a range?

i = 1    ## initialization Points
n = 4
sum = 0
while (i <= n):    ## Condition
    sum += i
    i += 1         ## Increment
    print(sum)
print("Total: ",sum)
```

```python
# Q4. 
i = 1
sum = 0
n = int(input("Enter range: "))
while i <= n:
    sum += i**2
    i += 1
print(sum)
```

```python
#Q5. Write a program to print the sum of cube of a given number and check the number is Armstrong or not.?
n = int(input("Enter Number: "))
og = n
sum = 0

while n > 0:
    digit = n % 10
    cube = digit ** 3
    sum += cube
    n = n // 10   

print("Sum of cubes:", sum)

if sum == og:
    print(og, "is an Armstrong number.")
else:
    print(og, "is not an Armstrong number.")
```

```python
#Q6.
n = int(input("Enter Number: "))
even = 0
odd = 0
while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even += digit
    else:
        odd += digit
    n //= 10
print("The Sum of even  numbers in a digit:", even)
print("The Sum of odd  numbers in a digit:", odd)
```

```python
#Q7. Reverse of a Number.

n =int(input("Enter Number: "))
reverse = 0
while n > 0:
    digit = n % 10
    reverse = 10 * reverse + digit
    n //= 10
print(reverse)
```

```python
#Q8. #Palindrome Check
num = int(input("Enter Number: "))
reverse = 0
originalnumber = num

while num > 0:
    digit = num % 10
    reverse = 10 * reverse + digit
    num //= 10
print("The Reverse of the number is: ", reverse)

if reverse == originalnumber:
    print("It is a Palindrome")
else:
    print("It is not a Palindrome")
```

```python
#Q12. STR= "Pyhton Program,"
str1 = "Python Program"
str1 = str1.lower() 

vowels = "aeiou"
vowel_count = 0
consonant_count = 0

i = 0
while i < len(str1):
    ch = str1[i]
    if ch.isalpha():      
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1
    i += 1

print("Number of vowels:", vowel_count)
print("Number of consonants:", consonant_count)
```
