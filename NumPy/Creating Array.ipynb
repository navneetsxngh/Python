## Creating Arrays

NumPy is used to work with arrays. 
The array object in NumPy is called ndarray.

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr)

print(type(arr))
```

```python
## We can use a tuple to create a NumPy array.

arr = np.array((1,2,3,4,5))
print(arr)
```

## Dimension in Array

A Dimension in arrays is one level of array depth(nested arrays).

```python
## 0 Dimension Array
arr = np.array(42)
print(arr)

np.ndim(arr)
```

```python
## 1 Dimension Array
arr = np.array([1,2,3,4,5])
print(arr)

np.ndim(arr)
```

```python
## 2 Dimension array

arr = np.array([[1,2,3],[4,5,6]])
print(arr)

np.ndim(arr)
```

```python
## 3 Dimension Array
arr = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])
print(arr)

np.ndim(arr)
```

```python
a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```

```python
## N dimentional array
arr = np.array([1,2,3,4], ndmin= 5)
print(arr)

np.ndim(arr)
```
