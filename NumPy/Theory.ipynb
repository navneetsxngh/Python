```python
# Pre-Defined Function in Numpy
# (1). zeros() ---> It will create an array in which all the values will be zero in either 1-D or multi-dimentional

h = np.zeros(3)
print(h)

i = np.zeros((3,4))
print(i)
```

```python
# (2). ones()---> It will create an array in which all the values will be 1 in either 1-D or multi-dimentional
j = np.ones(4)
print(j)
print("\n")
k = np.ones((4,5))
print(k)
```

```python
# (3). eye() ---> It will create an array in which diagonal positional elements will be 1 and rest all are 0.
# Syntax : eye(rows, columns)
a = np.eye(3)                 ## Symmetric Matrix
print(a)          
print("\n")
b = np.eye(3,4)               ## Asymmetric Matrix  ---> rows != columns
print(b)
```

```python
# (4). diag() ---> It will create an array in which we can set our custom values at diagonal position.
a = np.diag([43,34,67,90])
print(a)
```

```python
# (5). Random Module
# (a). randint() ---> It will create an array randomly in a given range.
# syntax : np.random.randint(min_range, max_range, total_numbers)

a = np.random.randint(1,10,3)
print(a)
```

```python
# (b). rand() ---> It will generate random numbers from 0 - 1.
# syntax : np.random.rand(total_random_number)

a = np.random.rand(4)
print(a)
```

```python
## Suppose we have
# dataset(emp_data)(emp-id, salary, domain) ---> randomly column add (working_hours)
```

```python
# (c). seed() ---> It will fix or lock our generated data.
# syntax : np.random.seed(45)
# a = np.random.randint(1,10,2)

np.random.seed(34)
a = np.random.randint(1,10,3)
print(a)

## Now our value is locked with the seed value '45'.
```

```python
# (d). view and copy
a = np.array([10,20,30,40,50,60,70,80,90])
print(a)

b = a[1:4]
b[:] = 0
print(b)
print(a)
```

```python
## Copy
a = np.array([10,20,30,40,50,60,70,80,90])
b = a[1:4].copy()
b[:] = 0
print(b)
print(a)
```

```python
# Unique ---> It will create an array in which it will remove all the duplicate values and return 3 arrays
## arr ----> Unique elements of array
## return_index = True ---> Indexing of unique elements
## return_counts = True ---> Frequency of each unique elements.
```

```python
a = np.array([1,2,3,1,1,1,1,4,5,4,4,4,4,4,4,6,7])
print(a)
```

```python
b = np.unique(a, return_index = True, return_counts = True)
print(b)
```

```python
# linspace() ---> It will return an array in which elements will be same gap in a given range.
# Syntax : np.linspace(min_num, max_num, total_numbers)

a = np.linspace(1,2,5)
print(a)
```

```python
## Horizontal and Vertical Stacking
# Horizontal Stacking ---> It will add 2 or more arrays horizontally.

a = np.arange(1,5)
b = np.arange(5,9)
c = np.arange(9,13)

print(np.hstack((a,b,c)))
print("\n")
print(np.vstack((a,b,c)))
```
