```python
#Two Important -----> Object and Class
#Class ---> It is basically the blueprint of the object (Template of the object)
#Object ---> Real world Entity

## We use "Class" keyword to create a class.

#Here we are creating a class having attributes (name, location and branch)
class  student:
    
    # This (__init__) function automatically called when the object is created.
    def __init__(self, name, location, branch):      #Constructor --> It creates a object and allocates the memory for it.
        self.name = name
        self.location = location
        self.branch = branch

    #Displaying all the properties of the student.
    def display_info(self):
        print(f"Name : {self.name}")
        print(f"Location : {self.location}")
        print(f"Branch : {self.branch}")

# Here we are creating the object
student1 = student("Rahul", "Jaipur", "ESE")
student2 = student("Shyam", "Jaipur", "CSE")

print(student1.name)
print(student2.location)
print(student1.branch)
print("\n")
student1.display_info()
```

```python
# Make class of car (wheels, color, engine, bulletproof, sunroof, brand, model)
class car:
    def __init__(self, wheels, color, engine, bulletproof, sunroof, brand, model):
        self.wheels = wheels
        self.color = color
        self.engine = engine
        self.bulletproof = bulletproof
        self.sunroof = sunroof
        self.brand = brand
        self.model = model
        
    def display_info(self):
        print(f"Brand : {self.brand}")
        print(f"Wheels Company : {self.wheels}")
        print(f"Color : {self.color}")
        print(f"Engine : {self.engine}")
        print(f"Bullet Proofing : {self.bulletproof}")
        print(f"Sunroof : {self.sunroof}")
        print(f"Model : {self.model}")

virtus = car("Apollo", "Midnight", "1500Cc", True, True, "Volkswagen", "Virtus")
virtus.display_info()
```
