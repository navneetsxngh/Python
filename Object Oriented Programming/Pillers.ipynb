```python
## Inheritance
class Person:     ## This is a Parent Class
    def __init__(self, name, id, gender):
        self.name = name
        self.id = id
        self.gender = gender

    def display_info(self):
        print(f"Name : {self.name}")
        print(f"Identity : {self.id}")
        print(f"Gender : {self.gender}")

# Inheriting all the attributes from the parent class to the child class
class Student(Person):
    def __init__(self, name, id, gender, marks):
        super().__init__(name, id, gender)
        self.marks = marks

    def display_info(self):
        super().display_info()
        print(f"Marks : {self.marks}")

class Teacher(Person):
    def __init__(self, name, id, gender, salary, department):
        super().__init__(name, id, gender)
        self.salary = salary
        self.department = department
    
    def display_info(self):
        super().display_info()
        print(f"Salary : {self.salary}")
        print(f"Department : {self.department}")


print("------Student Deatils------")
student = Student("Navneet", 101, "Male", 95)
student.display_info()
print("\n------Teacher Deatils------")
teacher = Teacher("Vinay", 100, "Male", 100000, "CSE")
teacher.display_info()
```

```python
class Animals:
    def __init__(self, type, gender):
        self.type = type
        self.gender = gender
    def display_animal(self):
        print(f"Animal : {self.type}")
        print(f"Gender : {self.gender}")

class Category(Animals):
    def __init__(self, type, gender, sound):
        super().__init__(type, gender)
        self.sound = sound
    def display(self):
        super().display_animal()
        print(f"Sound : {self.sound}")

animal1 = Category("Dog", "Male", "Woof")
animal2 = Category("Cat", "Female", "Meow")

animal1.display()
print("\n----------------\n")
animal2.display()
```

## Multi-Level Inheritance

```python
class GrandFather:
    def message(self):
        print("This Message From Grandfather")

#Inheriting the property of the GrandFather to the Father
class Father(GrandFather):
    def message_from_father(self):
        print("This Message from Father")

#Inheriting the property of Father to Child
class Child(Father):
    def message_from_child(self):
        print("This Message from Child")

# Creating the child object
c = Child()

#This will Show Message from grandfather
c.message()

#This will show message from Father
c.message_from_father()

# This will show message from child
c.message_from_child()
```

## Multiple Inheritance

```python
class Father:
    def manner(self):
        return "Has Good Manners"

class Mother:
    def bargaining(self):
        return "Has Good Bargaining Skills!"

class Child(Mother, Father):
    def gaming(self):
        return "Has Good Gaming Skills!"

c = Child()

print(c.manner())
print(c.bargaining())
print(c.gaming())
```

```python
class Teacher:
    def __init__(self, subject, qualification, experience):
        self.subject = subject
        self.qualification = qualification
        self.experience = experience
    
    def teach(self):
        print(f"Teaches : {self.subject}")
        print(f"Qualifications : {self.qualification}")
        print(f"Experience : {self.experience}")

class Singer:
    def __init__(self, songs, genre, albums_released):
        self.songs = songs
        self.genre = genre
        self.albums_released = albums_released
    
    def sing(self):
        print(f"Songs : {self.songs}")
        print(f"Genre : {self.genre}")
        print(f"Albums Released : {self.albums_released}")

class Person(Teacher, Singer):
    def __init__(self,name, age, city, subject, qualification, experience, songs, genre, albums_released):
        Singer.__init__(self, songs, genre, albums_released)
        Teacher.__init__(self, subject, qualification, experience)
        self.name = name
        self.age = age
        self.city = city
    
    def display(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"City : {self.city}")

p = Person("Navneet", 24, "Jaipur", "Physics", "Post Graduate", "5 Years", "Not Afraid", "Rap", "Kamikaze")
p.display()
p.teach()
p.sing()
```

## Polymorphism

It is a concept in which functions have many forms.

- Overloading
- Overriding

```python
class A:
    def show(self):
        print("Hello From A")
class B(A):
    def show(self):
        print("Hello From B")
class C(B):
    def show(self):
        print("Hello From C")
class D(C):
    pass

obj = D()

# for obj in [A(), B(), C()]:
#     obj.show()

D.__mro__  #Method Resolution Order ---> It defines the order in which methods are searched
                                        #  for and inherited in a class hierarchy especially crucial in 
                                        # scenarios involving multiple inheritance.
```

```python
class Mother:
    def skills(self):
        print("Managing House")
class Father:
    def skills(self):
        print("Managing Finances")
class Child(Mother, Father):
    pass

obj1 = Child()
Child.__mro__
obj1.skills()
```

#### Result Calculation 

```python
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    def display(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_no}")


class Marks(Student):
    def __init__(self, name, roll_no, physics, chemistry, maths):
        super().__init__(name, roll_no)
        self.physics = physics
        self.chemistry = chemistry
        self.maths = maths

    def display(self):
        super().display()
        print(f"Marks in Physics: {self.physics}")
        print(f"Marks in Chemistry: {self.chemistry}")
        print(f"Marks in Maths: {self.maths}")

class Result(Marks):
    def __init__(self, name, roll_no, physics, chemistry, maths):
        super().__init__(name, roll_no, physics, chemistry, maths)
        self.total = physics + chemistry + maths
        self.average = self.total / 3

    def display(self):
        super().display()
        print(f"Total Marks: {self.total}")
        print(f"Average Marks: {self.average:.2f}")

student1 = Result("Navneet Singh", 101, 85, 90, 88)
student1.display()
```

#### Shape Area

```python
import math

class Shape:
    def area(self):
        print("Area cannot be determined for generic shape.")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        area = math.pi * self.radius ** 2
        print(f"Area of Circle is {area:.2f}")

class Rectangle(Shape):
    def __init__(self, width, length):
        self.width = width
        self.length = length
    
    def area(self):
        area = self.width * self.length
        print(f"The area of Rectangle is {area}")

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        area = 0.5 * self.base * self.height
        print(f"The area of Triangle is {area}")

s = [Circle(5), Rectangle(5, 10), Triangle(10, 6)]

for i in s:
    i.area()
```

#### Bank Account

```python
class Account:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{self.name} deposited ₹{amount}. New balance: ₹{self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"{self.name} - Insufficient balance!")
        else:
            self.balance -= amount
            print(f"{self.name} withdrew ₹{amount}. New balance: ₹{self.balance}")

class SavingsAccount(Account):
    def __init__(self, name, balance=0, interest_rate=0.03):
        super().__init__(name, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"{self.name} earned ₹{interest:.2f} interest. New balance: ₹{self.balance:.2f}")

    def withdraw(self, amount):
        if self.balance - amount < 500:
            print(f"{self.name} - Minimum balance of ₹500 required!")
        else:
            super().withdraw(amount)

class CurrentAccount(Account):
    def __init__(self, name, balance=0, overdraft_limit=1000):
        super().__init__(name, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > self.balance + self.overdraft_limit:
            print(f"{self.name} - Overdraft limit exceeded!")
        else:
            self.balance -= amount
            print(f"{self.name} withdrew ₹{amount}. New balance: ₹{self.balance}")

savings = SavingsAccount("Riya", 2000)
current = CurrentAccount("Amit", 1000)


savings.deposit(500)
current.deposit(800)

savings.withdraw(1800)    
savings.withdraw(1000)    
current.withdraw(1700)    
current.withdraw(2500)    

savings.add_interest()
```

## Teacher-Student Relationship

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")

class Teacher(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary
    
    def display(self):
        super().display()
        print(f"Salary : {self.salary}")

class Student(Person):
    def __init__(self, name, age, marks):
        super().__init__(name, age)
        self.marks = marks
    
    def display(self):
        super().display()
        print(f"Marks Obtained : {self.marks}")

teacher1 = Teacher("Mr. Sharma", 40, 55000)
student1 = Student("Navneet", 24, 92)

people = [teacher1, student1]

for person in people:
    print(f"---Displaying Result for : {(person)}")
    person.display()
    print("\n")
```

## Payment System

```python
class Payment:
    def pay(self):
        print("Amount Payable")

class CreditCardPayment(Payment):
    def pay(self):
        super().pay()
        print("Pay through Credit Card")

class PayPalPayment(Payment):
    def pay(self):
        super().pay()
        print("Paid Through PayPal")

class UPIPayment(Payment):
    def pay(self):
        super().pay()
        print("Payment Through UPI")

payments = [CreditCardPayment(), PayPalPayment(), UPIPayment()]

for payment in payments:
    print("\n--- Payment Mode ---")
    payment.pay()
```
