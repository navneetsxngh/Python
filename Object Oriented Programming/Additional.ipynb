```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def display(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"Gender : {self.gender}")

class Singer(Person):
    def __init__(self, name, age, gender, songs, genre, albums_released):
        super().__init__(name, age, gender)
        self.songs = songs
        self.genre = genre
        self.albums_released = albums_released
    def display(self):
        super().display()
        print(f"Total Songs : {self.songs}")
        print(f"Genre : {self.genre}")
        print(f"Albums Released : {self.albums_released}")

class Teacher(Person):
    def __init__(self, name, age, gender, experience, qualifications):
        super().__init__(name, age, gender)
        self.experience = experience
        self.qualifications = qualifications
    def display(self):
        super().display()
        print(f"Experience : {self.experience}")
        print(f"Qualifications : {self.qualifications}")

teacher = Teacher("Harshita", 20, "Female", "5 Years", "Bachelor of Science")
singer = Singer("Seedhe Maut", 30, "Male", 100, "Rap", 10)

teacher.display()
print("\n")
singer.display()
```

```python
class TeacherRole: # Renamed Teacher to TeacherRole to emphasize it's a component
    def _init_(self, subject, qualification, experience):
        self.subject = subject
        self.qualification = qualification
        self.experience = experience
    
    def teach(self):
        """Prints teaching details."""
        print("\n--- Teaching Role ---")
        print(f"Teaches : {self.subject}")
        print(f"Qualifications : {self.qualification}")
        print(f"Experience : {self.experience}")

class SingerRole: # Renamed Singer to SingerRole
    def _init_(self, songs, genre, albums_released):
        self.songs = songs
        self.genre = genre
        self.albums_released = albums_released
    
    def sing(self):
        """Prints singing details."""
        print("\n--- Singing Role ---")
        print(f"Songs : {self.songs}")
        print(f"Genre : {self.genre}")
        print(f"Albums Released : {self.albums_released}")

class Person:
    """
    A person who HAS a TeacherRole and/or a SingerRole (Composition).
    Initialization is clean, only accepting role data optionally.
    """
    def _init_(self, name, age, city, teacher_data=None, singer_data=None):
        self.name = name
        self.age = age
        self.city = city
        
        # Initialize roles as None
        self.teacher_role = None
        self.singer_role = None

        # Create Role Objects ONLY if data is provided
        if teacher_data:
            # The 'TeacherRole' class takes arguments from the dictionary
            self.teacher_role = TeacherRole(**teacher_data)

        if singer_data:
            self.singer_role = SingerRole(**singer_data)
    
    def display(self):
        print("--- Personal Details ---")
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"City : {self.city}")

    # The Person class now delegates the action to the role object
    def teach(self):
        if self.teacher_role:
            self.teacher_role.teach()
        else:
            print("\n(This person has no defined teaching role.)")

    def sing(self):
        if self.singer_role:
            self.singer_role.sing()
        else:
            print("\n(This person has no defined singing role.)")

# =========================================================================
# EXAMPLE 1: Creating a Person who is ONLY a Teacher (Clean Initialization)
print("--- Example 1: Teacher Only (Passing a Dictionary) ---")
p_teacher = Person(
    name="Rupali Sharma", 
    age=30, 
    city="Pune",
    teacher_data={
        "subject": "History", 
        "qualification": "M.A.", 
        "experience": "8 Years"
    }
    # singer_data is omitted, defaulting to None
)

p_teacher.display()
p_teacher.teach()
p_teacher.sing() 

print("\n" + "="*50 + "\n")

# EXAMPLE 2: Creating a Person who is ONLY a Singer (Clean Initialization)
print("--- Example 2: Singer Only (Passing a Dictionary) ---")
p_singer = Person(
    name="Manish Verma", 
    age=24, 
    city="Jaipur",
    singer_data={
        "songs": "Not Afraid", 
        "genre": "Rap", 
        "albums_released": "Kamikaze"
    }
    # teacher_data is omitted, defaulting to None
)

p_singer.display()
p_singer.teach() 
p_singer.sing()
```

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance
    ## Getter to get the Balance
    def display_balance(self):
        print(f"Your Balance is : {self.__balance}")
    
    ## Setter to Deposit the Balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Amount Deposited Successfully")
            self.display_balance()
    
    ## Withdrawing the amount
    def withdraw(self, amount):
        if amount < self.__balance:
            self.__balance -= amount
            print("Amount Withdrawn Successfully")
            self.display_balance()

acc = BankAccount("Devendra", 1000000)
acc.deposit(500000)
acc.withdraw(600000)
```
