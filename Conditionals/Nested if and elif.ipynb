### Nested if and elif 

#### Temperature Check

```python
temp = float(input("Enter Temperature: "))

if temp == 25:
    print("Normal")
elif temp < 25:
    print("Cold")
else:
    print("Hot")
```

#### Generate Discounted Bill

```python
amount = int(input("Enter Amount (in Rs): "))

if amount <= 1000:
    print("10% Discount")
    disamount = 10/100 * amount
    amount_to_pay = amount - disamount
    print("Amount to pay = ", amount_to_pay)
elif amount > 1000 and amount <= 5000:
    print("15% Discount")
    disamount = 15/100 * amount
    amount_to_pay = amount - disamount
    print("Amount to pay = ", amount_to_pay)
elif amount > 5000 and amount <= 10000:
    print("20% Discount")
    disamount = 20/100 * amount
    amount_to_pay = amount - disamount
    print("Amount to pay = ", amount_to_pay)
else:
    print("25% Discount")
    disamount = 25/100 * amount
    amount_to_pay = amount - disamount
    print("Amount to pay = ", amount_to_pay)
```

#### Digit in Words

```python
digit = int(input("Enter Digit: "))

if digit == 0:
    print("Zero")
elif digit == 1:
    print("One")
elif digit == 2:
    print("Two")
elif digit == 3:
    print("Three")
elif digit == 4:
    print("Four")
elif digit == 5:
    print("Five")
elif digit == 6:
    print("Six")
elif digit == 7:
    print("Seven")
elif digit == 8:
    print("Eight")
elif digit == 9:
    print("Nine")
else:
    print("Invalid Digit Input")
```

#### Leap Year Or Not

```python
year = int(input("Enter Year: "))

if year % 100 == 0:
    if year % 400 == 0:
        print("Leap Year")
    else:
        print("Not a Leap Year")
elif year % 4 == 0:
    print("Leap Year")
else:
    print("Not a leap Year")
```

```python
year = int(input('Enter Year: '))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```
