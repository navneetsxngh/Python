## If - Else Statement

```python
a = 10
if a > 9:
  print("Positive")
  print("Bye")
#print("Hello")
```

```python
data = 10
if data != 9:
    print("Hello")
if data == 10:
    print("Bye")
```

```python
num = int(input("Enter Number: "))
if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
data = list(input("Enter Number: "))
print(data)
```

```python
num = list(input("Enter number: "))    #It stores string value in the list.
if '9' in num:
    print("Present")
else:
    print("Absent")
```

```python
num1 = int(input("Enter Number: "))
num2 = int(input("Enter Number: "))
num3 = int(input("Enter Number: "))

if num1 > num2 and num1 > num3:
    print(num1)
elif num2 > num1 and num2 > num3:
    print(num2)
elif num3 > num2 and num3 > num1:
    print(num3)
```

```python
d1 = [1,2]          #List is mutable 
d2 = [1,2]
if d1 == d2:     #d1[0] == d2[0]  and d1[1] == d2[1]
    print("Equal")
else:
    print("Not Equal")
```

```python
a = int(input("Enter Number: "))
b = int(input("Enter Number: "))
c = input("")

if '+' in c:
    print(a+b)
elif '-' in c:
    print(a-b)
elif '*' in c:
    print(a*b)
elif '/' in c:
    print(a/b)
else:
    print("Invalid Operation")
```
