# *EDA ==> Explorartory Data Analysis*

## Parts of EDA
1.) Univariate Analysis ==> Analysis on a single independent column

2.) Bivariate Analysis ==>  Analysis on two columns.

3.) Multivariate Analysis ==> Analysis on more than 2 columns.

## Data Types
1.) Numerical Data ==> Continous Data ==> Age(year, month, days), Height

2.) Categorical Data ==> Discrete Data ==> Total Number of Employees, Martial Status

```python
df['Survived'].value_counts()
```

```python
## To see the data in table we use Bar
df['Survived'].value_counts().plot(kind = 'bar')
```

```python
## If we have to find out percentage we use piechart
df['Survived'].value_counts().plot(kind = 'pie', autopct = '%.2f')            ## autopct = Auto Percentage
```

```python
## If we have Numerical data then we use Histogram
# because it finds the distribution.

plt.hist(x = df['Age'])
plt.show()
```

```python
## Boxplot ---> to find out outliers.

# 1. Lower fence
# 2. 25% data
# 3. IQR (Inter Quartile Range)(75% - 25%)
# 4. 75% data
# 5. Upper fence
```
