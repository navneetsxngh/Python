```python
# (1). Series ---> It is one-dimentional array and it returns only values not column name.
a = pd.Series([1,45,78,90])
print(a)
print(type(a))
```

```python
# (2). DataFrame ---> It is Multi-dimentional array and it returns values with column name.
## Task :: We have to create a dataframe of employees in which column are 
# emp_id, emp_name, department, salary, working_hours
```

```python
a = {
    "Emp_ID" : ['A101', 'A102', 'A103', 'A104', 'A105', 'A106'],
    "Emp_Name" : ['Sam', 'Rahul', 'Aniket', 'Gourav', 'Mohit', 'Karan'],
    "Department" : ['IT', 'Operations', 'IT', 'IT', 'Operations', 'IT'],
    "Salary" : [20000, 25000, 35000, 40000, 21000, 45000],
    "Working_Hours" : [8,8,9,9,8,9]
}
df = pd.DataFrame(a)
print(df)
```

```python
b = {
    "Emp_ID" : [1,2,3,4],
    "Name" : ['Navneet', 'Rahil', 'Faruk', 'Manish'],
    'Department' : ['IT', 'Travel', 'Travel', 'IT'],
    'Salary' : [120000, 150000, 250000, 50000]
}
df = pd.DataFrame(b)
print(df)
```

```python
# ===============================================
# PANDAS DATA ANALYSIS: INITIAL INSPECTION
# Based on W3Schools "Pandas - Analyzing DataFrames"
# ===============================================

# 1. Setup and Data Loading
# -----------------------------------------------

import io

# Create a sample CSV data string for demonstration (simulating the 'data.csv' file)
csv_data = """Duration,Pulse,Maxpulse,Calories
60,110,130,409.1
60,117,145,479.0
60,103,135,340.0
45,109,175,282.4
45,117,148,400.0
60,102,127,300.0
120,101,126,300.0
60,120,132,250.0
45,104,134,253.3
60,98,124,269.0
45,123,178,
120,100,120,200.0
"""

# Use io.StringIO to read the string data as if it were a file
df = pd.read_csv(io.StringIO(csv_data))

print("DataFrame successfully loaded.")

# 2. Viewing the Data: head() and tail()
# -----------------------------------------------

# **Markdown Cell:** Quick Visual Inspection

# The `head()` method is used to get a quick overview of the DataFrame by displaying the top rows.

# **Code Cell:** df.head(n)

print("--- Displaying the Top 5 Rows (df.head()) ---")
print(df.head())

# **Markdown Cell:** Checking the End of Data

# The `tail()` method is used to view the last rows, which can be important for checking footers or recent entries.

# **Code Cell:** df.tail(n)

print("\n--- Displaying the Last 3 Rows (df.tail(3)) ---")
print(df.tail(3))


# 3. Info About the Data: info()
# -----------------------------------------------

# **Markdown Cell:** Detailed Summary and Null Check

# The `info()` method provides a summary of the DataFrame structure, including data types and, critically, non-null counts for each column.

# **Code Cell:** df.info()

print("\n--- Displaying Data Information (df.info()) ---")
df.info()

# **Markdown Cell:** Result Interpretation

# The output of `df.info()` is essential for identifying data quality issues:
# 1. **RangeIndex (12 entries):** Tells us there are 12 rows in total.
# 2. **Dtype:** Shows the data type (e.g., 'float64', 'int64').
# 3. **Non-Null Count:** For the 'Calories' column, the count is 11, which is less than the total 12 entries. This indicates **1 missing value (Null Value)**, signaling a need for data cleaning.
```
