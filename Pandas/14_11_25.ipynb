```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

## Bivariate Analysis

### 1.) scatterplot(Numerical column, numnerical column)

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'])           #df['total_bill']
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data=tips, hue = tips['sex'])
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue= tips['sex'], style= tips['smoker'])
```

### 2.) HeatMap(Categorical - Categorical)

```python
p = pd.crosstab(tips['day'], tips['time'])
p
```

```python
sns.heatmap(p)
```
