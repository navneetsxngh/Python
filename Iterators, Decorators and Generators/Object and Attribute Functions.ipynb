## 1.) type(object, bases = None, dict = None)

```python
print(type(10))
print(type(10.235))
print(type("Hello"))
print(type([1,2,3]))
print(type((1,2,3)))
print(type({1,2,3}))
print(type({"a": 1}))
print(type(None))
```

## 2.) isinstance(object, classinfo, /)

```python
## It checks whether the object belongs to a specified class.
x = 10
print(isinstance(x, int))
print(isinstance(x, float))
print(isinstance(x, (int, str)))

print(isinstance("Hello", (int, str)))
```

## 3.) hasattr(object, attribute)

```python
## Checking if a string has a specified Method.
text = "Hello"

hasattr(text, "lower")
hasattr(text, "search")
```

## 4.) getattr(object, attribute, default)

```python
import math
getattr(math, 'pi')
getattr(math, 'sqrt')(25)
```

## 5.) id(object, /)

```python
x = 10
y = 10
print(id(x))
print(id(y))   ## Same ID's but

l1 = [1,2,3,4,5]
l2 = [1,2,3,4,5]
print(id(l1))
print(id(l2))   ## Different ID's
```

## 6.) dir(object = None, /)

```python
dir(list)
#It gives all the methods of list class
dir(math)  #--> All functions is math module.
```

## 7.) repr(object)

```python
text = "Hello World"
print(text)
print(repr(text))
```
