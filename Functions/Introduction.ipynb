```python
# Q1 Create a function who add sum of square of a given number.
def sumsquare(a):
    sum1 = 0
    while a > 0:
        digit = a % 10       
        square = digit ** 2 
        sum1 += square
        a //= 10            
    print("Sum of Square:", sum1)

sumsquare(121)
```

```python
# Q2 Create a Function who check given number is a Armstrong or not.?

def armstrong(a):
    sum2  = 0
    original = a
    while a > 0:
        digit = a % 10
        cube = digit ** 3
        sum2 += cube
        a //= 10
    if original == sum2:
        print("Armstrong")
    else:
        print("Not Armstrong")

armstrong(153)
```

```python
# Q3 Create a Function who check given number is a palindrome or not.?

def palindrome(a):
    sum3 = 0
    original = a
    while a > 0:
        digit = a % 10
        sum3 = sum3 * 10 + digit
        a //= 10
    if sum3 == original:
        print("Palindrome")
    else:
        print("Not Palindrome")

palindrome(35468465)
```

```python
# Q4 Write a function to count the digits in a number.?

def count(a):
    counter = 0
    while a > 0:
        a //= 10
        counter += 1
    print("Number of Digits: ", counter)

count(12345)
```

```python
# Q5 Write a function to count the frequency of each element in a list.?
def frequency():
    a = [1,2,2,3,3,3,4]
    d = {x:a.count(x) for x in a}
    print("Frequency: ", d)

frequency()
```

```python
#Q7 Write a function who reverse a string without using pre-defined function and slicing.

def reverse_string(a):
    for i in a:
        print(a[::-1])
        break

text = "Programming"

reverse_string(text)
```

```python
#Q8 Write a function who calculate second maximum number in a list without using predefined function.
def second_maximum(a):
    max1 = a[0]
    max2 = a[0]

    for i in a:
        if i > max1:
            max2 = max1
            max1 = i
        elif i > max2 and i != max1:
            max2 = i
    print("Second Maximum number in a list is : ", max2)


leest = [1,2,3,4,5,6,7,8,9]
second_maximum(leest)
```

```python
#Q9 Write a function who adds digits of a number.
# input = 123
# output = 6

def digitadd(a):
    sum = 0
    while a > 0:
        digit = a % 10
        sum += digit
        a //= 10
    print("Sum of digit: ", sum)

c = 123
digitadd(c)
```

```python
#Q10 Write a function who create user-defined list and we can add, remove and update element.

d = [2, 3, 4, 5, 6]

def add_element(a):
    val = int(input("Enter value to add: "))
    a.append(val)
    print("List after adding:", a)

def remove_element(a):
    val = int(input("Enter value to remove: "))
    if val in a:
        a.remove(val)
        print("List after removing:", a)
    else:
        print("Value not found in list!")

def update_element(a):
    index = int(input("Enter index to update: "))
    if 0 <= index < len(a):
        new_val = int(input("Enter new element: "))
        a[index] = new_val
        print("List after updating:", a)
    else:
        print("Invalid index!")

add_element(d)
remove_element(d)
update_element(d)
```

```python
#Q11 Write a function to find longest word in a sentence.

#input = "Python Makes Coding Intresting"

a = "Python Makes Coding Intresting"

def longest_word(a):

    b = a.split()
    longest = len(b[0])

    for i in b:
        if len(i) > longest:
            longest = len(i)
    for i in b:
        if longest == len(i):
            print(i)

longest_word(a)
```

```python
#Q12 Write a function who remove vowels in a string.

a = "PYTHON PROGRAMMING"

def rmvowels(a):
    vowels = 'aeiou'
    b = ""
    for i in a.lower():
        if i not in vowels:
            b += i
    print(b)

rmvowels(a)
```

```python
#Q13 Write a function which determines the mean in a list.
a = [10,20,30,40,50,60]

def mean(a):
    sum1 = 0
    for i in a:
        sum1 += i
        mean = sum1 / len(a)
    print(mean)

mean(a)
```
